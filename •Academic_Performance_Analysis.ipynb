```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/Expanded_data_with_more_features.csv')
df
```

```python
df.info()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df
```

```python
df.drop_duplicates()
```

```python
df['Gender'].isnull()
```

```python
plt.figure(figsize=(10,5),facecolor='orange')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='Gender')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('Gender')
```

We analayse from students data,The heatmap shows that female students scored higher in Reading (73) and Writing (73) compared to males (66 and 64, respectively), whereas males performed better in Math (69) than females (64).

```python
plt.figure(figsize=(10,5),facecolor='blue')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='WklyStudyHours')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('WeelyStudyHours')
```

we can analyse from student data,Students who study more than 10 hours per week have the highest scores across all subjects (Math: 69, Reading: 70, Writing: 70), while those studying less than 5 hours per week have the lowest scores (Math: 65, Reading: 68, Writing: 67). The middle category (5-10 hours) shows moderate performance,

```python
plt.figure(figsize=(10,5),facecolor='green')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='TestPrep')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('Test-preparation',color='green')
```

We can analyse from student data,Students who completed test preparation scored significantly higher in all subjects (Math: 70, Reading: 74, Writing: 75) compared to those who did not (Math: 65, Reading: 67, Writing: 65).

```python
plt.figure(figsize=(10,5),facecolor='yellow')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='ParentEduc')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('Parent-Education',color='r')
```

We can analyse from students data,Students whose parents have a master's degree scored the highest across all subjects (Math: 72, Reading: 76, Writing: 76), while those with "some high school" education had the lowest scores (Math: 63, Reading: 66, Writing: 64).

```python
plt.figure(figsize=(10,5),facecolor='grey')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='EthnicGroup')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('EthnicGroup',color='black')
```

We can analyse from student data,The scores range from 73 to 75, with Group E performing the best across all subjects (Math: 75, Reading: 74, Writing: 73) and Groups A and B having the lowest scores (Math: 63).

```python
plt.figure(figsize=(10,5),facecolor='purple')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='ParentMaritalStatus')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('ParentMaritalStatus',color='black')
```

We can analyse from student data,Students with divorced, married, and widowed parents scored similarly in Math (67), while students with single parents had a slightly lower Math score (66). Reading scores range between 69 and 70, with students from divorced and widowed backgrounds performing best. Writing scores show slight differences, with divorced and widowed students scoring highest (69).

```python
plt.figure(figsize=(10,5),facecolor='yellow')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='NrSiblings')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('Siblings',color='black')
```

We can analyse from student data,The scores for Reading and Writing seem higher and more consistent compared to Math across all sibling categories.
Students with 7 siblings performed relatively better in Math and Reading than those with fewer siblings.



```python
plt.figure(figsize=(10,5),facecolor='turquoise')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='IsFirstChild')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('FirstChild',color='black')
```

We can analyse from student data,IsFirstChild tend to have slightly higher scores across all subjects, with an average Math score of 67, Reading score of 70, and Writing score of 69. Non-firstborn students perform lower in Math (66) and Reading (69) but show a similar Writing score (68).

```python
plt.figure(figsize=(10,5),facecolor='olive')
dff=df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],index='LunchType')
sns.heatmap(dff,annot=True,linewidth=0.2)
plt.title('Lunch',color='black')
```

We can analyse from student data,Students reduced lunch have significantly lower scores (Math: 59, Reading: 64, Writing: 63) compared to those with standard lunch (Math: 71, Reading: 72, Writing: 72).
